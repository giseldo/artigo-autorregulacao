```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('..\data\questionario_assertivas.csv', delimiter=';')
df.head(3)
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(), cmap='coolwarm', vmin=-1, vmax=1)
```

```python
correlacao = df.corr()
correlacao_filtrada = correlacao[(correlacao >= 0.5) | (correlacao <= -0.5)]
sns.heatmap(correlacao_filtrada, cmap='coolwarm', vmin=-1, vmax=1)
```
