```python

#medias = medias.sort_values(ascending=False)
X = medias.index
Y = medias.values
desvio_padrao = stds.values

fig, ax = plt.subplots()
ax.barh(X, Y, xerr=desvio_padrao,capsize=5, color='skyblue', edgecolor='black')
ax.set_title("Perfil de autorregulao da amostra")
ax.set_ylabel("Constructo")
ax.set_xlabel("Mdia (%)")

#linhas_verticais = range(50, 110, 10)
#for linha in linhas_verticais:
#    ax.axvline(x=linha, color='red', linestyle='--', linewidth=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Dados de exemplo
data = {
    'Categoria': medias.index,
    'Valores': medias*100/7,
    'Desvio_Padrao': desvio_padrao*100/7
}
df = pd.DataFrame(data)

# Ordenar o DataFrame pelo valor
df_sorted = df.sort_values(by='Valores', ascending=True)

# Configurando os dados a partir do DataFrame ordenado
categorias = df_sorted['Categoria']
valores = df_sorted['Valores']
desvios_padrao = df_sorted['Desvio_Padrao']

# Criando o grfico de barras horizontal
plt.figure(figsize=(10, 6))
plt.barh(categorias, valores, xerr=desvios_padrao, capsize=5, color='skyblue', alpha=0.7)

# Adicionando rtulos e ttulo
plt.ylabel('Constructo')
plt.xlabel('Mdia Respostas')
plt.title('Perfil de autorregulao da amostra com Desvio Padro (Ordenado)')
#plt.xlim(2.5, 8)
#ax.set_title("")
#ax.set_ylabel("Constructo")
#ax.set_xlabel("Mdia (%)")


# Invertendo a ordem das categorias para que a maior fique no topo
plt.gca().invert_yaxis()

# Mostrando o grfico
plt.show()
```
