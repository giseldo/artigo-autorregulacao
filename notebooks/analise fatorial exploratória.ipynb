```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import FactorAnalysis
import numpy as np

data = pd.read_csv("questionario_respostas.csv", delimiter=";")

# Estatísticas descritivas
estatisticas_descritivas = data.describe()

# Análise Fatorial Exploratória (AFE)
# Normalizar os dados (opcional, dependendo dos dados)
data_normalized = (data - data.mean()) / data.std()

# Ajustar o modelo de análise fatorial
fa = FactorAnalysis(n_components=5, random_state=0)  # Tentaremos extrair 5 fatores
fa.fit(data_normalized)

# Cargas fatoriais
cargas_fatoriais = pd.DataFrame(fa.components_.T, index=data.columns, columns=[f'Fator {i+1}' for i in range(5)])
cargas_fatoriais
```

```python
import pandas as pd
import numpy as np
from factor_analyzer import FactorAnalyzer, calculate_bartlett_sphericity, calculate_kmo
import matplotlib.pyplot as plt

df = pd.read_csv("questionario_respostas_p.csv", delimiter=";")

chi_square_value, p_value = calculate_bartlett_sphericity(df)
print(f'Teste de esfericidade de Bartlett: chi2={chi_square_value}, p={p_value}')

kmo_all, kmo_model = calculate_kmo(df)
print(f'Medida KMO: {kmo_model}')

fa = FactorAnalyzer()
fa.fit(df)

ev, v = fa.get_eigenvalues()
print(f'Eigenvalues: {ev}')

plt.scatter(range(1, df.shape[1] + 1), ev)
plt.plot(range(1, df.shape[1] + 1), ev)
plt.title('Scree Plot')
plt.xlabel('Fatores')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()

fa = FactorAnalyzer(n_factors=5, rotation='varimax')
fa.fit(df)

# Exibir as cargas fatoriais
factor_loadings = fa.loadings_
#print(f'Cargas fatoriais:\n{factor_loadings}')

# Exibir variância explicada por fator
variance = fa.get_factor_variance()
#print(f'Variância explicada por fator:\n{variance}')

# Passo 4: Interpretar os fatores
# Imprimir as cargas fatoriais de forma mais legível
cargas_fatoriais = pd.DataFrame(factor_loadings, index=df.columns, columns=['Fator 1', 'Fator 2', 'Fator 3', 'Fator 4', 'Fator 5'])
#print(f'Cargas fatoriais:\n{cargas_fatoriais}')

# Passo 5: Exibir variáveis carregadas em cada fator
limiar = 0.4  # Defina o limiar de carga fatorial

# Função para encontrar variáveis que carregam em cada fator
def variaveis_por_fator(loadings, threshold):
    fatores = {}
    for i in range(loadings.shape[1]):
        fator_name = f'Fator {i+1}'
        fatores[fator_name] = loadings.index[abs(loadings.iloc[:, i]) >= threshold].tolist()
    return fatores

# Obter variáveis carregadas em cada fator
variaveis_fatores = variaveis_por_fator(cargas_fatoriais, limiar)
for fator, variaveis in variaveis_fatores.items():
    print(f'{fator}',"=~", end="")
    for var in variaveis:
        print(f'  {var}', "+", end="")
    print("", end="\n")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from factor_analyzer import FactorAnalyzer, calculate_bartlett_sphericity, calculate_kmo


df = pd.read_csv("questionario_respostas_p.csv", delimiter=";")

# Limpeza dos dados e normalização
df_cleaned = df.dropna().astype(float)
data_scaled = StandardScaler().fit_transform(df_cleaned)

# Testes de esfericidade de Bartlett e KMO
chi_square_value, p_value = calculate_bartlett_sphericity(data_scaled)
kmo_all, kmo_model = calculate_kmo(data_scaled)

print(f"Chi-Square Value: {chi_square_value}, p-value: {p_value}")
print(f"KMO Model: {kmo_model}")

# Determinar o número de fatores
fa = FactorAnalyzer(n_factors=22, rotation=None)
fa.fit(data_scaled)

# Extrair os eigenvalues
ev, v = fa.get_eigenvalues()

# Plotar os eigenvalues
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(range(1, df_cleaned.shape[1]+1), ev, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()

# Escolher o número de fatores baseado no scree plot e eigenvalues
n_factors = sum(ev > 1)  # Critério de Kaiser

# Realizar a análise fatorial com o número escolhido de fatores
fa = FactorAnalyzer(n_factors=n_factors, rotation='varimax')
fa.fit(data_scaled)

# Mostrar as cargas fatoriais
loadings = fa.loadings_
#print(loadings)

cargas_fatoriais = pd.DataFrame(loadings, index=df.columns, columns=['Fator 1', 'Fator 2', 'Fator 3', 'Fator 4', 'Fator 5', 'Fator 6', 'Fator 7', 'Fator 8', 'Fator 9', 'Fator 10', 'Fator 11', 'Fator 12', 'Fator 13', 'Fator 14', 'Fator 15', 'Fator 16', 'Fator 17', 'Fator 18', 'Fator 19', 'Fator 20', 'Fator 21', 'Fator 22'])

limiar = 0.4  

def variaveis_por_fator(loadings, threshold):
    fatores = {}
    for i in range(loadings.shape[1]):
        fator_name = f'Fator{i+1}'
        fatores[fator_name] = loadings.index[abs(loadings.iloc[:, i]) >= threshold].tolist()
    return fatores

variaveis_fatores = variaveis_por_fator(cargas_fatoriais, limiar)
for fator, variaveis in variaveis_fatores.items():
    print(f'{fator}',"=~ ", end="")
    for var in variaveis:
        print(f'{var}', "+ ", end="")
    print("", end="\n")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from factor_analyzer import FactorAnalyzer, calculate_bartlett_sphericity, calculate_kmo

# Carregar os dados
df = pd.read_csv("questionario_respostas_p.csv", delimiter=";")

# Limpeza dos dados e normalização
df_cleaned = df.dropna().astype(float)
data_scaled = StandardScaler().fit_transform(df_cleaned)

# Testes de esfericidade de Bartlett e KMO
chi_square_value, p_value = calculate_bartlett_sphericity(data_scaled)
kmo_all, kmo_model = calculate_kmo(data_scaled)

print(f"Chi-Square Value: {chi_square_value}, p-value: {p_value}")
print(f"KMO Model: {kmo_model}")

# Determinar o número de fatores
fa = FactorAnalyzer(n_factors=22, rotation=None)
fa.fit(data_scaled)

# Extrair os eigenvalues
ev, v = fa.get_eigenvalues()

# Plotar os eigenvalues
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(range(1, df_cleaned.shape[1]+1), ev, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()

# Escolher o número de fatores baseado no scree plot e eigenvalues
n_factors = sum(ev > 1)  # Critério de Kaiser

# Realizar a análise fatorial com o número escolhido de fatores
fa = FactorAnalyzer(n_factors=n_factors, rotation='varimax')
fa.fit(data_scaled)

# Mostrar as cargas fatoriais
loadings = fa.loadings_

# Exibir as variáveis que carregam fortemente em cada fator
threshold = 0.4  # Limite para considerar uma carga significativa
factors = {}

for i in range(n_factors):
    factors[f'Fator {i+1}'] = df_cleaned.columns[loadings[:, i] > threshold].tolist()

# Exibir as variáveis de cada fator
for factor, variables in factors.items():
    print(f"{factor}=~ {variables}")
```
